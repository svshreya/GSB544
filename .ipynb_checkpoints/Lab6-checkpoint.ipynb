
## Dataset: Baseball Players

```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import make_column_selector, ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
from plotnine import ggplot, aes, geom_point, labs, theme, theme_minimal, geom_bar, element_text
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the data
baseball = pd.read_csv("/content/Hitters.csv")

# Get rid of columns with mostly NaN values
good_cols = baseball.isna().sum() < 100
baseball = baseball.loc[:,good_cols]

# Drop other NAs
baseball = baseball.dropna()

baseball.head()
```


# Part I: Different Model Specs

### A. Regression without regularization


Create a pipeline that includes all the columns as predictors for Salary, and performs ordinary linear regression

```python
X = baseball.drop(["Salary"], axis = 1)
y = baseball["Salary"]


ct = ColumnTransformer(
  [
    ("dummify",OneHotEncoder(sparse_output = False, handle_unknown='ignore'),
    make_column_selector(dtype_include=object)),("standardize",StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lr_pipeline = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
)
```

Fit this pipeline to the full dataset, and interpret a few of the most important coefficients.



```python
lr_model = lr_pipeline.fit(X,y)


# Get the coefficients of the linear regression model
coefs_linear = lr_pipeline.named_steps["linear_regression"].coef_
coefs_linear
```

### *Interpreting the Coefficients*

```python
# Get the feature names after one-hot encoding and standardization
feature_names_transformed = (ct.named_transformers_['dummify'].get_feature_names_out().tolist() +
                             make_column_selector(dtype_include=np.number)(X))

# Mapping coefficients to features
feature_coef_mapping = dict(zip(feature_names_transformed, coefs_linear))

# Sorted coefficients from most negative to most positive for better interpretation
sorted_feature_coefs = sorted(feature_coef_mapping.items(), key=lambda item: item[1])

sorted_feature_coefs
```

The most significant coefficients after fitting the model with standardized numerical features and one-hot encoded categorical features are:

CRuns: 480.75 - According to this, we can say that career runs scored have the most positive impact on a player's salary. An increase in career runs achieved increases salary by 480.75.

CAtBat: -391.04 - According to this, we can say that career at-bats have the largest negative impact on a player's salary. Every career at-bat decreases salary by 391.04.

Hits: 337.83 - According to this, we can say that the number of hits is a significant predictor of salary, with more hits correlating with higher pay. Every hit increases salary by 337.83.

AtBat: -291.09 - At-bats in the current year negatively impact salary, every at-bat decreases salary by 291.04.

CRBI: 260.69 - Career runs batted in (RBI) are strongly positively related to salary, indicating that players who drive in more runs over their careers tend to earn more.

CWalks: -213.89 - Career walks have a negative association with salary in this model, which could suggest that players who walk often may not be as valued as those who hit for power or average, depending on the context within which this model was developed.

Walks: 135.07 - Interestingly, while career walks negatively impact salary, walks in the current year show a positive correlation with salary. This might reflect a valuation for on-base ability in the short term.

CHits: 86.69 - Career hits positively influence salary, reinforcing the value of a player's hitting ability over their career.

PutOuts: 78.76 - Defensive metrics such as putouts positively affect salary, indicating good defensive play is valued.

Division_E: 58.42, Division_W: -58.42 - These coefficients indicate that playing in the Eastern division is associated with a higher salary compared to the Western division, which is the reference group for this model.

Use cross-validation to estimate the MSE you would expect if you used this pipeline to predict 1989 salaries.

```python
scores = cross_val_score(lr_pipeline, X, y, cv = 5, scoring = 'neg_mean_squared_error')

-scores
```


### B. Ridge Regression

1. Create a pipeline that includes all the columns as predictors for Salary, and performs ordinary ridge regression

```python
X = baseball.drop(["Salary"], axis = 1)
y = baseball["Salary"]


ct = ColumnTransformer(
  [
    ("dummify",OneHotEncoder(sparse_output = False, handle_unknown='ignore'),
    make_column_selector(dtype_include=object)),("standardize",StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

ridge_pipeline = Pipeline([('preprocessor', ct),('regressor', Ridge())])
```

2. Use cross-validation to tune the lambda
 hyperparameter.

```python
ridge_pipeline.fit(X, y)

alpha_values = [0.001, 0.01, 0.1, 1, 10]

alpha_grid = {'regressor__alpha': alpha_values}

gscv = GridSearchCV(ridge_pipeline, alpha_grid, cv = 5, scoring='neg_mean_squared_error')

gscv_fitted = gscv.fit(X, y)
```

```python
results = -gscv_fitted.cv_results_['mean_test_score']

results
```

```python
# Get the coefficients of the linear regression model
coefs_ridge = ridge_pipeline.named_steps["regressor"].coef_
```

My interpretations of the most important coefficients obtained with this best alpha value, where lambda=1:

Hits: 296.65 - The number of hits is a strong positive predictor of salary, indicating that players with more hits are likely to have higher salaries.

AtBat: -270.69 - The number of at-bats has a significant negative association with salary in this model. This might suggest that players who have more at-bats but perhaps do not convert these into positive outcomes (like hits or runs) are not necessarily higher-paid.

CHits: 126.66 - Career hits continue to show a strong positive effect on a player's salary, reinforcing the value of hitting skill over a player's career.

Walks: 124.41 - Surprisingly, while career walks negatively influence salary, current season walks have a strong positive effect, which may reflect teams' valuing players who can get on base through walks in the short term.

PutOuts: 78.62 - Reflecting a player's defensive contributions, putouts show a moderately positive relationship with salary.

Divisional Coefficients (Division_W: -60.01 and Division_E: 60.01) - These show the difference in salary based on the division, with the Eastern division being associated with higher salaries than the Western division.

In each group, the Elastic Net model usually has a lower MSE than both the Ridge and Lasso models meaning that the Elastic Net model provides a better fit to the data than LASSO and Ridge.

The reason Elastic Net might "win" or have a lower MSE is because of how it combines penalties. It includes the flexibility of both Ridge and Lasso methods, which allows it to handle various types of data structures effectively.

When multiple correlated features are present, Elastic Net has the ability to distribute the coefficient values among these features, as opposed to allocating a non-zero coefficient to one and reducing the coefficients of the others to zero. This approach preserves the collective influence of the correlated variables within the model, rather than possibly overlooking it.

The MSEs for the 'Best Numeric' set are significantly higher than for the other feature sets, which suggests that the 'All Predictors' and 'Top 5 Numeric' might contain essential predictors that are not included in the 'Best Numeric' set.
