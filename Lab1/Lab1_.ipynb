**Part 4: What other geometries could you use here besides geom_point? Provide an actual plot of another geometry being used here and why itâ€™s good or bad for visualizing these data.**

Another geometry that can be used other than geom_point is the violin plot style-geometry which I used below. After viewing the dataset and our variables,'internet_users', 'gdp', and 'income', I think a scatterplot would be a better visualization of the data rather than the violin plot. The scatterplot is more suitable to visualize the relationship between GDP and internet users across countries, which are both continuous variables since, you can clearly see the trend and correlation in these two variables and since scatterplots are meant especially for this purpose and with added aesthetics like filled color for regions and size for population, they can help better represent multi-dimensional data. A violinplot is not suitable if we wanted it to effectively, provide insight into the relationship between two continuous variables. Also compared to a scatterplot, 2 or more variables are less intuitively displayed and represented than in a scatterplot.


```python
warnings.filterwarnings('ignore')

#Creating the violinplot and storing it in a object
violin_plot3 = (p9.ggplot(df_q3, p9.aes(x='internet_users', y='gdp', fill ='four_regions', size = 'income'))
    + p9.geom_violin(alpha = 0.5, width = 5)

    # Annotations specified to add text to the geometry
    + p9.annotate("text", x = 105, y = 40, label ="\% of population", color = "grey", size = 10, ha = "right", va = "center")
    + p9.annotate("text", x = 0, y = 40000, label = "US$, inflation-adjusted", color ="grey", size = 10, ha = "right", angle = 90)
    + p9.annotate("text", x = 48, y = 3200, label = "2001", color = "grey", size = 140, alpha = 0.2)

    # Labels, Themes, Guide and color specified to enhance visualization
    + p9.labs(x='Individuals using the Internet', y='GDP/capita', fill='World Regions')
    + p9.theme_bw()
    + p9.theme(text = p9.element_text(size = 12), figure_size = (12, 6))
    + p9.scale_fill_manual(values = {"africa": "aqua", "asia": "salmon", "americas": "lawngreen", "europe": "yellow"},
                           labels = {"africa": "Africa", "asia": "Asia", "americas": "Americas", "europe": "Europe"})

    # Breaks and Labels specified for each axis
    + p9.scale_x_continuous(breaks = [10, 20, 30, 40, 50, 60, 70, 80, 90],
                            labels = ["10", "20", "30", "40", "50", "60", "70", "80", "90"],
                            limits = (0,110))

    + p9.scale_y_continuous(breaks = [200, 500, 1000, 2000, 5000, 10000, 20000, 50000,100000],
                            labels = ["200", "500", "1000", "2000", "5000", "10k", "20k", "50k", "100k"],
                            trans = 'log'))
#print/display violinplot created
print(violin_plot3)
```
