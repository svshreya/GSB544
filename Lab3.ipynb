# Function 1: pluralize_gift()

```python
obj_1 = "foot"
obj_2 = "baby"
obj_3 = "tree"

obj_1.find("oo")
obj_2[-1]
obj_3.find("oo")

obj_1.replace("oo", "ee")
obj_2.replace("y", "ies")
obj_3 + "s"
```

```python
def pluralize_gift(gift):
    """
    Returns plural of a noun
  
    Parameters
    ----------
    gift: str
      A noun
    
    Return
    ------
    str
    Plural version
    """
    if "oo" in gift:
        gift = gift.replace("oo", "ee")
    elif gift.endswith('y'):
        gift = gift[:-1] + 'ies'
    else:
        gift = gift + 's'

    return gift
```

# Function 2: make_phrase()

```python
def make_phrase(num, num_word, item, verb, adjective, location):
    """
    This function constructs a phrase based on the given inputs.
    
    Parameters:
    - num (int): The number associated with the item.
    - num_word (str): The word representation of the number.
    - item (str): The main gift item.
    - verb (str): The verb associated with the item, if any.
    - adjective (str): The adjective describing the item, if any.
    - location (str): The location associated with the item, if any.
    
    Returns:
    - str: The constructed phrase.
    """
  
    # Step 1: Replace NAs with blank strings
    if pd.isna(verb):
        verb = ""
    if pd.isna(adjective):
        adjective = ""
    if pd.isna(location):
        location = ""
  
  # Step 2: If the day number is larger than 1, the gift items need pluralized!
  # Hint: call the function you created above!
  
    if num > 1:
        item = pluralize_gift(item)
  # Step 3: Figure out if a gift item starts with a vowel
    starts_with_vowel = item[0].lower() in ['a', 'e', 'i', 'o', 'u']
    num_to_word = {
    1: 'one',
    2: 'two',
    3: 'three',
    4: 'four',
    5: 'five',
    6: 'six',
    7: 'seven',
    8: 'eight',
    9: 'nine',
    10: 'ten',
    11: 'eleven',
    12: 'twelve',
    }
    xmas['Num_Word'] = xmas['Day'].map(num_to_word)
    xmas2['Num_Word'] = xmas['Day'].map(num_to_word)

  # Step 4: For the first day, if the gift item starts with a vowel, replace the day with "an" and if the gift item does not start with a vowel, replace the day with "a" (e.g. a partridge in a pear tree). If it is not the first day, use just the number word (e.g. ten lords a leap)
    if num  == 1:
    
        # For the first day
        if starts_with_vowel:
            num_word = "an"
        else:
            num_word = "a"
    else:
        num_word = num_to_word.get(num)
  # Step 5: Put all of the pieces together into one string and return!
    
    phrase = f"{num_word} {adjective} {item} {verb} {location}"
    phrase = ' '.join(phrase.split())
    return phrase  
```

# Function 3: sing_day()

```python
def sing_day(dataset, num, phrase_col):
    """
    This function sings the '12 Days of Christmas' song for a specified day, 
    using the given dataset and phrase column. It ensures that the song's lines 
    are grammatically correct with proper use of commas, "and", and periods.
    
    Parameters:
    - dataset (pd.DataFrame): The dataset containing song phrases and day info.
    - num (int): The day number (e.g., 1 for the first day).
    - phrase_col (str): The name of the column in the dataset containing the phrases for each day.
    
    Returns:
    - str: The song for the specified day, with each gift on its own line and grammatically correct punctuation.
    """
    
    # Setup the intro line
    num_word = dataset['Day.in.Words'].iloc[num-1]
    intro = f"On the {num_word} day of Christmas, my true love gave to me:"
    
    # Extract the gift phrases
    gifts = dataset[phrase_col].iloc[:num].tolist()
    
    # Adjust for 'and' and add commas at the end of each gift except for the last one
    if num > 1:
        gifts[0] = "and " + gifts[0]
    
    # Combine the gifts in reverse order
    song_gifts = ",\n".join(gifts[::-1])
    
    # Combine intro and gifts and add a period at the end
    song = f"{intro}\n{song_gifts}."
    
    return song
```
