## Starbucks

```python
response = requests.get("https://www.menuism.com/restaurant-locations/starbucks-coffee-39564")


# Parse the HTML using BeautifulSoup
soup_starbucks = BeautifulSoup(response.content, 'html.parser')

# Find all the list items containing Starbucks Coffee locations
locations_list = soup_starbucks.find_all("div", attrs={"class":"card popular-cities-box"})[0:]

```

```python
state_locations = []

for location in locations_list:
    location_names = location.find_all("li", attrs={"class":""})  # Use find_all here
    for location_name in location_names:
        state_name = location_name.get_text(strip=True)

        # Split the text at " Starbucks Coffee locations " and parentheses
        expression = re.split(r' Starbucks Coffee locations \((\d+)\)', state_name)

        # The first part is the state name, and the second part is the number of locations (if the string format is consistent)
        state = expression[0].strip()  # (The strip() is used to remove any leading/trailing white spaces)
        num_locations = int(expression[1]) if len(expression) > 1 else None  # Convert the string to an integer
        state_locations.append({"State":state,"Number of Locations":num_locations})

pd.DataFrame(state_locations)
```

# Automate

Convert your code for Exercises 1-3 above to a function that takes a single argument: the menuism URL. This function should

Scrape the information on state names and corresponding number of store locations on the webpage specified

Extract the name of the company from either the URL specified or the webpage

Return a clean, organized dataset that has three columns: state abbreviation, location count, company name. Company name will likely be repeated on every row.

```python
def locations_by_state(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract coffee shop name from the <h3> tag
    h3_text = soup.find('h3').get_text(strip=True)
    match = re.search(r'\d+ (.+?) Locations', h3_text)
    if match:
        coffee_shop_name = match.group(1)
    else:
        coffee_shop_name = "Unknown Shop"  # Default name if not found in HTML

    # Find all the list items containing restaurant locations
    locations_list = soup.find_all("div", attrs={"class":"card popular-cities-box"})

    state_locations = []
    for location in locations_list:
        location_names = location.find_all("li", attrs={"class":""})
        for location_name in location_names:
            state_name_text = location_name.get_text(strip=True)

            # Using regex to extract state name and number of locations
            match = re.match(r'(\w+).*\((\d+)\)', state_name_text)
            if match:
                state, num_locations = match.groups()
                state_abbr = stateabb(state)

                state_locations.append({
                    "Company Name": coffee_shop_name,
                    "State": state_abbr,
                    "Number of Locations": int(num_locations)
                })

    # Convert the state_locations list to a  DataFrame
    df = pd.DataFrame(state_locations)
    return df
```
