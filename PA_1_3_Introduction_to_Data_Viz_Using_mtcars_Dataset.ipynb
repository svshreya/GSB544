# Data Visualization Using `plotnine`

The Palmer Penguins dataset had a variety of column types with which to explore some nice visualizations. The `mtcars` dataset is another popular dataset for doing some simple data work, but does not contain the same types of variables.

Run the following code to load the `mtcars` dataset and explore the observations and variables contained within. To learn more about this dataset check out [this site](https://stat.ethz.ch/R-manual/R-devel/library/datasets/html/mtcars.html)

```python
from plotnine import ggplot, geom_point, aes, geom_boxplot, geom_histogram, geom_density, geom_point, geom_line, labs
import statsmodels.api as sm
import pandas as pd
import seaborn as sns
```

```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
```

```python
df
```

The statement "the `mtcars` dataset does not contain the same types of variables as the penguins dataset" is a little true and a little false. There are no variables that contain text values, BUT there are variables that it makes sense to consider categorical variables. In most situations, it makes sense to treat the values of categorical variables as text values instead of numeric values.

Use the `astype()` method to convert the categorical variables of the `mtcars` (df) dataset to have text values in the code chunk below. We've done one for you!

```python
df["am"] = df["am"].astype(str)

# Convert the other variables below
df["gear"] = df["gear"].astype(str)
df["carb"] = df["carb"].astype(str)
df["cyl"] = df["cyl"].astype(str)
df["vs"] = df["vs"].astype(str)
```

**What happens if you convert the am variable back to float values and then try to create this same plot? Explain the differences between the two plots.**

When the 'am' is in 'str' format it will be treated as a categorical variable. Every unique string value is seen as a separate category on the x-axis. When a boxplot of mpg miles per gallon by the am variable iss mad, we see two box plots side by side, one with cars that have automatic transmission (am="0") and one with cars that have manual transmission (am="1"). However, When am is converted back to a float type, the 'am' variable am is treated as a continuous variable. When a boxplot is made using plotnine, the am variable is plotted on a continuous scale, and it may not display accurate results as the x-axis variable is not categorical. Thus, this plot doesn't make intuitive sense.

```python
#before converting the 'cyl' and 'hp' values to float values, our histogram is as shown
plot = (
    ggplot(mtcars, aes(x ='hp', fill ='cyl'))
    + geom_histogram(alpha = 0.5, bins = 10)  # alpha is for transparency
)
plot
```

```python
df["cyl"] = df["cyl"].astype(float)
#df["hp"] = df["hp"].astype(float)
```

```python
#after converting the 'cyl' and 'hp' values to float values, our histogram is as shown
plot = (ggplot(mtcars, aes(x='hp', fill='cyl'))
    + geom_histogram(alpha=0.5))  # alpha is for transparency

plot
```

**What happens if you convert the cyl variable back to float values and then try to create this same plot? Explain the differences between the two plots.**


As explained similarly above, when we convert the 'cyl' variable to a string type, it's treated as a categorical variable by plotnine. When you create a histogram, the x-axis displays discrete bars for every unique string value for 4-cylinder, 6-cylinder, and 8-cylinder cars. Each bar's height is representative of how many records exist for that category. However, when the 'cyl' variable is converted to a 'float' value, it is treated as a continuous variable. Again, the histogram may not show a very intuitive representation of the data, as it is continuous data.

Some of these variables, like the `cyl` variable, have numeric values that actually make sense as numbers (i.e. the number of cylinders in the engine). However, it doesn't make the most sense to "do math" with this type of variable (e.g. take averages and such) because there are so few different value this can take on AND they're an explicit choice made by the car manufacturer. So, it makes more sense to treat the `cyl` variable as a categorical variable despite it having numeric values.
