```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from plotnine import ggplot, aes, geom_point, geom_smooth, theme_light, labs
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from palmerpenguins import load_penguins
penguins = load_penguins()
penguins.head()
```

```python
penguins = penguins.dropna()
X = penguins[['bill_length_mm', 'species', 'island', 'sex', 'flipper_length_mm', 'body_mass_g', 'year']]
y = penguins['bill_depth_mm']
X_train, X_test, y_train, y_test = train_test_split(X, y)

lr = LinearRegression()
enc = OneHotEncoder()

ct = ColumnTransformer(
    [("dummify_s", enc, ['species']),
    ("dummify_i", enc, ['island']),
    ("dummify_sex", enc, ['sex'])]
)

my_pipeline = Pipeline(
    [('dummify_everything', ct), ('ols', lr)]
)

pipeline1 = my_pipeline.fit(X_train, y_train)
pipeline1
```

# Best Multiple Linear Model

```python
from sklearn import pipeline
penguins = penguins.dropna()
X = penguins[['bill_length_mm', 'species', 'island', 'sex', 'flipper_length_mm', 'body_mass_g', 'year']]
y = penguins['bill_depth_mm']
X_train, X_test, y_train, y_test = train_test_split(X, y)

lr = LinearRegression()
enc = OneHotEncoder()

ct = ColumnTransformer([
    ("scale", StandardScaler(), ['bill_length_mm', 'flipper_length_mm', 'body_mass_g']),
    ("dummify_s", enc, ['species']),
    ("dummify_i", enc, ['island']),
    ("dummify_sex", enc, ['sex'])
], remainder='passthrough')

pipeline = Pipeline(
    [('dummify_everything', ct), ('ols', lr)])

best_linear_model = pipeline.fit(X_train, y_train)
best_linear_model
```

```python
# making predictions with both training and testing data
y_pred_train = my_pipeline.predict(X_train)
y_pred_test = my_pipeline.predict(X_test)

# find the mean squared error
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)
```

```python
scores = cross_val_score(best_linear_model, X, y, cv = 5, scoring = 'neg_mean_squared_error')
score1 = -scores.mean()
score1
```

# KNN Models

```python
X = penguins[['bill_length_mm', 'species', 'island', 'sex', 'flipper_length_mm', 'body_mass_g', 'year']]
y = penguins['bill_depth_mm']

X_train, X_test, y_train, y_test = train_test_split(X, y)

ct = ColumnTransformer([
    ("scale", StandardScaler(), ['bill_length_mm', 'flipper_length_mm', 'body_mass_g']),
    ("dummify_s", enc, ['species']),
    ("dummify_i", enc, ['island']),
    ("dummify_sex", enc, ['sex'])
], remainder='passthrough')

knn_model3 = Pipeline(steps=[("ct", ct), ("knn", KNeighborsRegressor(n_neighbors=3))])

knn_model3.fit(X_train, y_train)

# making predictions with both training and testing data
knn_model3_y_pred_train = knn_model3.predict(X_train)
knn_model3_y_pred_test = knn_model3.predict(X_test)

# find the mean squared error
knn_model3_mse_train = mean_squared_error(y_train,knn_model3_y_pred_train)
knn_model3_mse_test = mean_squared_error(y_test, knn_model3_y_pred_test)
```

```python
knn_model7 = Pipeline(steps=[("ct", ct), ("knn", KNeighborsRegressor(n_neighbors=7))])

knn_model7.fit(X_train, y_train)


# making predictions with both training and testing data
knn_model7_y_pred_train = knn_model7.predict(X_train)
knn_model7_y_pred_test = knn_model7.predict(X_test)

# find the mean squared error
knn_model7_mse_train = mean_squared_error(y_train, knn_model7_y_pred_train)
knn_model7_mse_test = mean_squared_error(y_test, knn_model7_y_pred_test)
```

# Decision Tree Model

```python
# Create the Decision Tree model pipeline
dt_pipeline = Pipeline(steps=[
    ('transform', ct),
    ('regressor', DecisionTreeRegressor())
])

# Fit the pipeline to the training data
dt_pipeline.fit(X_train, y_train)
```

```python
# making predictions with both training and testing data
dt_model_y_pred_train = dt_pipeline.predict(X_train)
dt_model_y_pred_test = dt_pipeline.predict(X_test)

# find the mean squared error
dt_model_mse_train = mean_squared_error(y_train, dt_model_y_pred_train)
dt_model_mse_test = mean_squared_error(y_test, dt_model_y_pred_test)
```

```python
scores = cross_val_score(dt_pipeline, X, y, cv = 5, scoring = 'neg_mean_squared_error')

score4 = -scores.mean()
score4
```
